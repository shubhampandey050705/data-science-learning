```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset=pd.read_csv(r"C:\Users\Shubham Pandey\Downloads\Social_Network_Ads (1).csv")
dataset.head()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
sns.scatterplot(x=dataset["Age"],y=dataset["EstimatedSalary"],hue=dataset["Purchased"])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x)
x=pd.DataFrame(scaler.transform(x), columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
knn.score(x_test,y_test)*100
```

```python
knn.score(x_train,y_train)*100
```

```python
for i in range(1,20):
    knn1=KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i, knn1.score(x_test,y_test)*100, knn1.score(x_train,y_train)*100) 
```
