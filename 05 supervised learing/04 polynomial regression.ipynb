```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Shubham Pandey\Downloads\salary.csv")
dataset
```

```python
dataset.columns = dataset.columns.str.strip()
```

```python
plt.figure(figsize=(10,4 ))
plt.scatter(dataset["Level"], dataset["Salary"], color='blue', alpha=0.5)
plt.show()
```

```python
dataset.corr() #highly correlated data
```

```python
x=dataset[["Level"]]
y=dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x=pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test , y_test)*100 #accuracy is very pure hence the dataset is not sutaible for this model type
```

```python
predicted_salary = lr.predict(x)
```

```python
plt.figure(figsize=(10,4 ))
plt.scatter(dataset["Level"], dataset["Salary"], color='blue', alpha=0.5)
plt.plot(dataset ["Level"], predicted_salary, color='red')
plt.title("Salary vs Level")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(["Predicted Salary", "Actual Salary"])
plt.show() #from the graph we can see that the predicted salary is not close to the actual salary, so we need to improve the model
```
