```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset =pd.read_csv(r"C:\Users\Shubham Pandey\Downloads\placement (2).csv")
dataset
```

```python
sns.scatterplot(x="cgpa", y="resume_score", data=dataset, hue="placed")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2 , random_state=42)
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train , y_train)
lr.score(x_test, y_test)*100
```

```python
lr.predict([[8.5, 80]])
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr, legend=2)
plt.show()
```

logistic regression with poynomial function


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Set random seed for reproducibility
np.random.seed(42)

# Generate 2D feature data
n_samples = 500
X = np.random.uniform(-3, 3, (n_samples, 2))

# Define a non-linear (polynomial) decision boundary: x2 > x1²
y = (X[:, 1] > (X[:, 0] ** 2 - 1)).astype(int)

# Visualize the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Polynomial Decision Boundary Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset= pd.read_csv(r"C:\Users\Shubham Pandey\Downloads\iris.csv")
dataset
```

```python
dataset["variety"].unique()
```

```python
sns.pairplot(data= dataset, hue="variety")
```

```python
x= dataset.iloc[:,:-1]
y=dataset["variety"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2 , random_state=42)
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression(multi_class="ovr")
lr.fit(x_train , y_train)
lr.score(x_test, y_test)*100
```

```python
lr.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
lr1= LogisticRegression()
lr1.fit(x_train , y_train)
lr1.score(x_test, y_test)*100
```

```python
lr1= LogisticRegression(multi_class="multinomial")
lr1.fit(x_train , y_train)
lr1.score(x_test, y_test)*100
```
