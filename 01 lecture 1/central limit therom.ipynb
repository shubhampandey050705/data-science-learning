```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_data =[np.random.randint(10,1000) for i in range(10000)] # between 10 and 1000 generated random population data 10000 times
population_data
```

```python
np.mean(population_data) # mean of the population data
```

```python
population_table = pd.DataFrame({"population": population_data})
population_table
```

```python
np.mean(population_data) # mean of the population data
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="population" ,data=population_table)   #this data is not normally distributed now we normalize it using CLT
plt.title("Population Density Distribution")
```

```python
sample_mean=[]
for num_sample in range(70):
    sample_data=[]
    for data in range(500):
        sample_data.append(np.random.choice(population_data))
        sample_mean.append(np.mean(sample_data))
      
```

```python
np.mean(sample_mean)
```

```python
sample_M = pd.DataFrame({"sample_mean": sample_mean})
plt.figure(figsize=(4,3))
sns.kdeplot(x="sample_mean", data=sample_M)  # this data is normally distributed now
plt.title("Sample Mean Density Distribution")
plt.show()
```
