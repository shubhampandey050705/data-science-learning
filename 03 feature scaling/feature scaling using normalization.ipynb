```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset=pd.read_csv(r"C:\Users\Shubham Pandey\Downloads\loan.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
ms= MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_minMaxScale"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.distplot(dataset["CoapplicantIncome"], color='blue')
plt.title("Original Coapplicant Income Distribution")
plt.subplot(1, 2, 2)
sns.distplot(dataset["CoapplicantIncome_minMaxScale"], color='green')
plt.title("Min-Max Scaled Coapplicant Income Distribution")
plt.show()
```
