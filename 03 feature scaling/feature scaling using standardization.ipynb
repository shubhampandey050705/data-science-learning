```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv(r"C:\Users\Shubham Pandey\Downloads\loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
ss=StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]= pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]), columns=["ApplicantIncome_ss"])
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["ApplicantIncome"], color="blue")

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["ApplicantIncome_ss"], color="red")

plt.show()
```
